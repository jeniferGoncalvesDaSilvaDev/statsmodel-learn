```python
import pandas as pd
import numpy as np
import statsmodels.api as st
import statsmodels.formula.api as stf
```

```python
data = st.datasets.get_rdataset("Guerry","HistData").data
data
```

```python
data.head(6)
```

```python

# Fit OLS regression use ln
# With R style formulas
res = stf.ols("Lottery ~ Literacy + np.log(Pop1831)",data=data).fit()
```

```python
res.summary()
```

```python
#usando array numpy
nobs =100
x = np.random.random((nobs,2))
x
```

```python
x =st.add_constant(x)
x
```

```python
beta = [1,.1,.5]
e = np.random.random(nobs)
y=np.dot(x,beta) +e
res_2 = st.OLS(y,x).fit()
res_2.summary()
```
